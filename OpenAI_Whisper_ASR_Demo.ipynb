```python

# Import whisper
import whisper

# Load the model
model = whisper.load_model("base")  # Change "base" to "tiny", "small", "medium", or "large" if needed

# Load audio and pad/trim it to fit 30 seconds
audio = whisper.load_audio("/content/User6.mp3")
audio = whisper.pad_or_trim(audio)

# Make log-Mel spectrogram and move to the same device as the model
mel = whisper.log_mel_spectrogram(audio).to(model.device)

# Detect the spoken language
_, probs = model.detect_language(mel)
print(f"Detected language: {max(probs, key=probs.get)}")

# Decode the audio
options = whisper.DecodingOptions()
result = whisper.decode(model, mel, options)

# Print the recognized text
print(result.text)

```

```python
import gradio as gr
import time
```

```python
def transcribe(audio):

    #time.sleep(3)
    # load audio and pad/trim it to fit 30 seconds
    audio = whisper.load_audio(audio)
    audio = whisper.pad_or_trim(audio)

    # make log-Mel spectrogram and move to the same device as the model
    mel = whisper.log_mel_spectrogram(audio).to(model.device)

    # detect the spoken language
    _, probs = model.detect_language(mel)
    print(f"Detected language: {max(probs, key=probs.get)}")

    # decode the audio
    options = whisper.DecodingOptions()
    result = whisper.decode(model, mel, options)
    return result.text
```

```python
gr.Interface(
    title="OpenAI Whisper ASR Gradio Web UI",
    fn=transcribe,
    inputs=gr.Audio(sources=["microphone"], type="filepath"),  # ✅ Fixed!
    outputs=gr.Textbox(),  # ✅ Corrected
    live=True
).launch()
```
